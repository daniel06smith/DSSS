```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# import data
data = "nfl-playbyplay_2009-2017_cleaned.csv"
df = pd.read_csv(data)
df.head()
```

As we work through this notebook, feel free to change different numbers and experiment to see how we can change our graphs easily using matplotlib and seaborn!

```python
# set the width, height of the figure
plt.figure(figsize=(16,6))
sns.lineplot(data=df, x="GameID", y="Yards.Gained")
```
